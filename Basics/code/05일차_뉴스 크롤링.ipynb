```python
import requests as req
from bs4 import BeautifulSoup
from datetime import datetime
import os
import pandas as pd

def get_newsdata(search) :
    url = 'https://search.naver.com/search.naver?where=news&query={}'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/100.0.48496.75'}

    # 검색할 개수(원하는 개수 / 10개 할것)
    count = 10
    title_list = []
    url_list = []
    news_list = []
    for i in range(count):
        res = req.get(url.format(search) + str(i), headers=headers).text
        soup = BeautifulSoup(res,"html.parser")
        title = soup.select(".news_area > a")
        news = soup.select(".info_group > a.info.press")

        for i in title :
            title_list.append(i.text)
            url_list.append(str(i.attrs["href"]))
        for i in news :
            news_list.append(i.text)
    data = {"기사제목" : title_list,"기사" : news_list, "url" : url_list}
    return data


# 검색어 검색
search = input("검색할 키워드 입력 :")
search = search.replace("","+")

data = get_newsdata(search)


# 데이터프레임 만들기
df = pd.DataFrame(data)
df
```
