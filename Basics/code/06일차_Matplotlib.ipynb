## 3. matplotlib 라이브러리

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import warnings
import matplotlib.font_manager as fm
warnings.filterwarnings(action="ignore")
```

### 3.1.1. 그래프 표현

```python
x = list(range(2005,2015))
y = [35,7,2,10,8,5,14,9,18,8]
plt.rcParams["font.family"] = "Malgun Gothic"

plt.plot(x,y, zorder = 2, color = "g") # 선
plt.bar(x,y, zorder = 1) # 막대
plt.scatter(x,y, zorder = 3) # 점
plt.title("연도")
plt.grid(True)
plt.show()
```

### 3.1.2. 다중 그래프 표현

```python
# 그래프 표현 공간 나누기
fig, ax = plt.subplots(2,2)
plt.rcParams['figure.figsize'] = (10,6)

# 축의 - 기호 표기
plt.rcParams["axes.unicode_minus"] = False

# 선 그래프
x = np.linspace(0,10,100)
y = np.cos(x)
ax[0,0].plot(x,y)


# 막대 그래프
x = np.arange(10)
y = np.random.uniform(1,10,10)
ax[0,1].bar(x,y)


# 점 그래프
x = np.random.rand(100)
y = np.random.randn(100)
ax[1,0].scatter(x,y)


# 이미지 그래프
z = np.random.uniform(0,1,(5,5))
ax[1,1].imshow(z)

plt.show()
```

#### - 1차 방정식 그래프

```python
plt.figure(figsize = (10,5))

#y = ax + b
a, b = 2,1
x = np.array([-5,0,10,14,19,20,25])
x2 = np.array([-5,0,10,14,19,20,25])
y = a*x + b
y2 = -a*x2 + b

plt.plot(x,y)
plt.plot(x2,y2)
plt.axvline(x=0,color = "black", linewidth = 0.5)
plt.axhline(y=0,color = "black", linewidth = 0.5)
plt.xticks(x)
plt.xlim(10,20) # x축 범위 지정
plt.ylim(-20,40) # y축 범위 지정
plt.show()
```

#### - 2차 방정식

```python
# 그래프 설정
plt.rc("font",family = "Malgun Gothic")
plt.rcParams["axes.unicode_minus"] = False


x = np.arange(-20,20,1)
a = 2
b = 1
c = 1
y = a*x**2 + b * x + c


plt.plot(x,y,"o", linestyle ="solid", c = "g")
plt.plot(x,a*x + c)
plt.plot(x,-b*x + c)
plt.axvline(x=0, color = "black", linewidth = 0.5)
plt.axhline(y=0, color = "black", linewidth = 0.5)
plt.title("$y = ax^2 + bx + c$")
# plt.fill_between(x[-2:2], y[:10])

plt.show()
```

#### - 이미지 안에 하이퍼링크 넣기2

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt


fig = plt.figure()
delta = 0.025
x = y = np.arange(-3.0, 3.0, delta)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 - Z2) * 2

im = plt.imshow(Z, interpolation='bilinear', cmap=cm.gray,
                origin='lower', extent=[-3, 3, -3, 3])

im.set_url('https://www.google.com/')
fig.savefig('../output/graph3.svg')
```

##### Quiz
- weather_total.csv를 읽어 기온-습도에 따른 구름(크기)과 일조시간(색상) 산점도 그래프 그리기

```python
data = pd.read_csv("../data/weather_total.csv", encoding = "cp949")
data

print(data.columns)
temperature = data["평균기온(°C)"]
humidity = data["평균 상대습도(%)"]
hour = data["합계 일조 시간(hr)"]
size = data["평균 전운량(1/10)"]
re_size = [s * 100 for s in size]

plt.scatter(temperature,humidity, c = hour, s = re_size, alpha =0.3)
plt.scatter(temperature,humidity, c = "b")
plt.xlabel("기온")
plt.ylabel("습도")
plt.title("기온-습도에 따른 구름(크기)과 일조시간(색상) 산점도")
plt.legend(hour)
plt.grid(True)
plt.show()
```
