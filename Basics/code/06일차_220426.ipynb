## 3. matplotlib 라이브러리

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings(action="ignore")
```

### 3.1 그래프 표현

```python
x = list(range(2005,2015))
y = [35,7,2,10,8,5,14,9,18,8]
plt.rcParams["font.family"] = "Malgun Gothic"

plt.plot(x,y, zorder = 3) # 선
plt.bar(x,y, zorder = 1) # 막대
plt.scatter(x,y, zorder = 2) # 점
plt.title("연도")
plt.grid(True)
plt.show()
```

### 2. 다중 그래프 표현

```python
# 그래프 표현 공간 나누기
fig, ax = plt.subplots(2,2)
plt.rcParams['figure.figsize'] = (10,6)

# 축의 - 기호 표기
plt.rcParams["axes.unicode_minus"] = False

# 선 그래프
x = np.linspace(0,10,100)
y = np.cos(x)
ax[0,0].plot(x,y)


# 막대 그래프
x = np.arange(10)
y = np.random.uniform(1,10,10)
ax[0,1].bar(x,y)


# 점 그래프
x = np.random.rand(100)
y = np.random.randn(100)
ax[1,0].scatter(x,y)


# 이미지 그래프
z = np.random.uniform(0,1,(5,5))
ax[1,1].imshow(z)

plt.show()
```

### 3. 파일 읽어서 그래프로 나타내기

```python
df = pd.read_csv("../data/HousingData.csv")
print(df.columns)

cols = ['CRIM','CHAS','NOX', 'RM','PTRATIO',"B"]
for idx,col in enumerate(cols):
    plt.subplot(2,3,idx+1)
    plt.scatter(df[col],df["MEDV"], s = 5)
    plt.xlabel(col)
    plt.ylabel("MEDV")
plt.show()
```

```python
df[df.columns].corr(method="pearson")
```

```python
import seaborn as sns
plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

fig = plt.figure(figsize = (10,6))
ax = fig.gca()

corr = df[df.columns].corr(method="pearson")

sns.heatmap(corr.values, annot = True,fmt = ".2f", xticklabels = df.columns, yticklabels = df.columns, ax=ax,
           cmap ="RdYlBu")
plt.tight_layout()
plt.show()
```
