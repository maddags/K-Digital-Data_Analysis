### 3.3.4 원 그래프

#### 기본 원 그래프

```python
# 데이터
idx = ["1학년","2학년","3학년","4학년","5학년","6학년"]
val = [16,12,7,5,10,8]
per = [ x/sum(val)*100 for x in val]

# 그래프
plt.pie(val, autopct = "%.1f%%") # %.3f%%, %.4f%% 해보기
plt.title("학년병 동아리 회원수")
plt.show()
```

#### - idx 텍스트 표기하기

```python
# 데이터
idx = ["1학년","2학년","3학년","4학년","5학년","6학년"]
val = [16,12,7,5,10,8]
per = [ x/sum(val)*100 for x in val]

# idx, 텍스트 표기
# plt.pie(per, autopct = "%1.1f%%", textprops = dict(color = "w"))
plt.pie(per, autopct = "%1.1f%%",labels = idx, textprops = {"color" : "b", "size" : 15})
plt.show()
```

#### - legend 위치 지정
- bbox_to_anchor=(left, bottom, width, height)

```python
# 데이터
idx = ["1학년","2학년","3학년","4학년","5학년","6학년"]
val = [16,12,7,5,10,8]
per = [ x/sum(val)*100 for x in val]

# 그래프
# plt.pie(val, autopct = "%.1f%%") # %.3f%%, %.4f%% 해보기
plt.pie(per, autopct = "%1.1f%%", textprops = dict(color = "w"))
plt.title("학년병 동아리 회원수")

# 범례
plt.legend(idx, title = "학년", loc = "center right",bbox_to_anchor=(0.65,0.2,0.5,1))


plt.show()
```

#### - 파이 조각이 분리되도록 나타내기

```python
# 도출 되는 정도 지정
explodes = [0.2,0.05,0.05,0.05,0.05,0.05]

plt.pie(per, autopct = "%1.1f%%", textprops = dict(color = "w"),
       explode = explodes)
plt.title("학년병 동아리 회원수")

plt.legend(idx, title = "학년", loc = "center right",bbox_to_anchor=(0.65,0.2,0.5,1))
plt.show()
```

#### - 그림자(shadow) 표기

```python
explodes = [0.2,0.05,0.05,0.05,0.05,0.05]

plt.pie(per, autopct = "%1.1f%%", textprops = dict(color = "w"),
       explode = explodes, shadow = True)
plt.title("학년병 동아리 회원수")

plt.legend(idx, title = "학년", loc = "center right",bbox_to_anchor=(0.65,0.2,0.5,1))
plt.show()
```

#### - angle 돌리기
- 양의 값 기준으로 시계 반대 방향으로 돌린다.

```python
explodes = [0.2,0.05,0.05,0.05,0.05,0.05]

plt.pie(per, autopct = "%1.1f%%", textprops = dict(color = "w"),
       explode = explodes, shadow = True,
       startangle = 30)
plt.title("학년병 동아리 회원수")

plt.legend(idx, title = "학년", loc = "center right",bbox_to_anchor=(0.65,0.2,0.5,1))
plt.show()
```

### 3.4.1 그래프의 열을 병합하여 나타내기

```python
grid = plt.GridSpec(2,3)
# grid = plt.Gridspec(2,3 wspace = 0.4, hspace = 0.3)

# 1행 선그래프
x = np.linspace(0,10,100) 
y = np.cos(x)
plt.subplot(grid[0,:2]).plot(x,y)

# 1행 이미지그래프
z = np.random.uniform(0,1,(5,5)) # 0~1 사이의 균일분포값을 5x5 형태로
plt.subplot(grid[0,2]).imshow(z)

# 2행 점 그래프
x = np.random.rand(100) # 정규 분포 데이터 랜덤 100개
y = np.random.rand(100)
plt.subplot(grid[1,0]).scatter(x,y)

# 2행 막대그래프
x = np.arange(10)
y = np.random.uniform(1,10,10) # 1~10 사이의 값 10개
plt.subplot(grid[1,1:]).bar(x,y)
plt.show()
```

#### - 히스토그램 
- 주어진 데이터를 몇개의 구간으로 나눠서 각 구간의 빈도수와 분포를 분석하는 그래프

```python
# 데이터
heights = np.array([175,165,164,164,171,165,160,169,164,159,
                    163,167,163,172,159,160,156,162,166,162,
                    158,167,160,161,156,172,168,165,165,177])

# 그래프
plt.hist(heights, bins = 6, edgecolor = "w") # bins : 구간 수
plt.title("히스토그램 그래프")
plt.xlabel("height")
plt.ylabel("도수")
plt.show()
```

#### - 누적 히스토그램

```python
heights = np.array([175,165,164,164,171,165,160,169,164,159,
                    163,167,163,172,159,160,156,162,166,162,
                    158,167,160,161,156,172,168,165,165,177])

plt.hist(heights, bins = 6, cumulative = True, edgecolor = "w", label = "cumulativeTrue")
plt.hist(heights, bins = 6, cumulative = False, edgecolor = "w", label = "cumulativeFalse")
plt.title("누적 히스토그램 그래프")
plt.xlabel("height")
plt.ylabel("빈도 수")
plt.legend()
plt.show()
```
