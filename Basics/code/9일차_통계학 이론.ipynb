#### 2.1.7 다변수함수(Multivariate Function)
- 복수의 입력변수를 갖는 함수 : z = f(x,y)
- 3차원 서피스 플롯(surface plot) or 컨투어 플롯(contour  plot)

```python
def f(x,y):
    return 2*x**2 + 6*x*y + 7*y**2 - 26*x - 54*y + 107

x = np.linspace(-3,7,100)
y = np.linspace(-3,7,100)
x,y = np.meshgrid(x,y)
z = f(x,y)

fig = plt.figure()
ax = fig.add_subplot(projection = "3d")
ax.plot_surface(x,y,z,linewidth = 0.1)
ax.view_init(40,-110)
plt.xlabel("$x$")
plt.ylabel("$y$")
ax.set_zlabel("$z$")
plt.title("Surface Plot")
plt.show()
```

#### 2.1.8 분리 가능 다변수 함수
- 다변수함수를 곱으로 표현 가능한 다변수 함수
- f(x,y) = f1(x)f2(y)
- 2차원 함수에 대한 1차원 단변수 함수는 3차원 공간산의 지형을 케이크 자르듯이 나눌 때 생기는 단면 모양과 같다. 그리고 분리 가능 다변수함수는 단면 모양이 모두 같다.
- $g(x,y) = exp(-x^2-16y^2) = exp(-x^2)exp(-16y^2) = gl(x)g2(y)$

```python
from matplotlib import transforms
from matplotlib.ticker import NullFormatter

plt.rcParams['font.family'] = 'Malgun Gothic'

def g1(x):
    return np.exp(-x ** 2)

def g2(y):
    return np.exp(-16 * y ** 2)

def g(x ,y):
    return g1(x) * g2(y)

xx = np.linspace(-1, 1, 100)
yy = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(xx, yy)
Z = g(X, Y)

left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left + width + 0.02
rect  = [left, bottom, width, height]
rectx = [left, bottom_h, width, 0.2]
recty = [left_h, bottom, 0.2, height]

plt.figure(1, figsize=(8,8))
ax = plt.axes(rect)
axx = plt.axes(rectx)
axy = plt.axes(recty)

nullfmt = NullFormatter()
axx.xaxis.set_major_formatter(nullfmt)
axy.yaxis.set_major_formatter(nullfmt)

ax.contour(X, Y, Z)
ax.axhline(0, c='r', ls='-')
ax.axhline(0.1, c='g', ls='--')
ax.axhline(0.2, c='b', ls='-.')
ax.axhline(0.3, c='c', ls=':')
ax.set_xlabel('x')
ax.set_ylabel('y')

axx.plot(xx, g1(xx), c='r')
axx.plot(xx, g2(0.1)*g1(xx), c='g', ls='--')
axx.plot(xx, g2(0.2)*g1(xx), c='b', ls='-.')
axx.plot(xx, g2(0.3)*g1(xx), c='c', ls=':')         
axx.set_title('g1(x)')
axx.text(-0.2, 0.3, 'y=0.3로 자른 단면')
axx.text(-0.2, 0.6, 'y=0.2로 자른 단면')
axx.text(-0.2, 0.9, 'y=0.1로 자른 단면')

base = axy.transData
rot = transforms.Affine2D().rotate_deg(-90)
axy.plot(yy, g2(yy), transform=rot + base)
axy.set_title('g2(y)')
axy.axhline(0, xmax=g2(0), c='r', ls='-')
plt.text(0.05, 0.02, '$g_2$(0)={:.2f}'.format(g2(0)))
axy.axhline(0.1, xmax=g2(0.1), c='g', ls='--')
plt.text(0.05, 0.12, '$g_2$(0.1)={:.2f}'.format(g2(0.1)))
axy.axhline(0.2, xmax=g2(0.2), c='b', ls='-.')
plt.text(0.05, 0.22, '$g_2$(0.2)={:.2f}'.format(g2(0.2)))
axy.axhline(0.3, xmax=g2(0.3), c='c', ls=':')
plt.text(0.05, 0.32, '$g_2$(0.3)={:.2f}'.format(g2(0.3)))

axx.set_xlim(ax.get_xlim())
axx.set_ylim(0, 1)
axy.set_xlim(ax.get_ylim())
axy.set_ylim(0, 1)

plt.suptitle('$g(x,y) = \exp(-x^2)\exp(-16y^2) = g_1(x)g_2(y)$의 그래프', y=1.04)
plt.show()
```

#### 2.1.9 다변수 다출력 함수(소프트맥스 함수)
- 입력변수뿐 아니라 출력변수도 여러 개 일 경우 출력을 벡터나 행렬로 나타내는 함수
- 소프트맥스(Sortmax) 함수는 다차원 벡터를 입력 받아서 다차원 벡터로 출력한다.
- 출력 벡터는 다음과 같은 특성이 있다.
> - 모든 출력 원소는 0,1 사잇값을 갖는다.
> - 모든 출력 원소의 합은 1이다.
> - 입력 원소의 크기 순서와 출력 원소의 크기 순서가 같다.
- 소프트맥스 함수는 다변수 입력을 확률처럼 보이게 출력한다.
>- 이러한 특성 때문에 인공신경망 마지막단계에서 출력을 조건부확률로 변경하는데 사용함

```python
def softmax(x,w) : # w : 가중치 벡터, x : 1차원 행렬
    e = np.exp(w * x)
    return np.exp(w*x) / e.sum()

x = [2.0,1.0,0.5]
y = softmax(x, np.ones(3))

print(f'x : {x}')
print(f'y : {y}')
print(f'sum(y) : {np.sum(y)}')

plt.plot(x,y)
plt.title("Softmax Function")
plt.show()
```

### 2.2 미분법

#### 데이터 분석에서 미분이 필요한 이유 

> [예제] 조절나사의 각도 x 결정하여 출력되는 음량 y를 재영하는 엠프가 있을 때 ---> 출력 음량(y)인 최대가 되는 조절나사의 각도(x)를 찾는 문제에서 미분이 사용됨 
> Y = wX + b 미분이 사용됨

- **데이터 분석 목표** : 예측 오차가 가장 적은 최적의 예측 모형을 구하는 것
- 예측 모형의 개수(모수)를 바꾸어 최적의 예측 모형을 찾는다. 이때 선능을 비교할 있어야 한다.
- 모수를 입력 받아서 오차(오차의 크기)를 출력하는 함수를 가장 적게 하는 것 --> 이런 함수를 목적함수(혹은 아래)라고 한다.
> - 손실함수(loss function)
> - 비용함수(cost function)
> - 오차함수(error function)
- 최적화 : 목적함수가 주어졌을 때 이 목적함수의 값을 가장 크게 또는 가장 작게 할 수 있는 입력값(모수)를 구하는 것
- **미분** : 입력값이 변했을 때 함수의 출력값이 어떻게 달라지는지를 정량적으로 알아야 하는 과정 

#### 기울기
- 변화율(y증감분 / x 증감분)에 따른 기울가(slope) 혹은 민감도(sensitivity)를 찾는 것
- 수치적 최적화는 가장 적은 횟수로 여러 x값을 시도하여 최적의 값을 찾는 방법

##### 3차 함수의 그래프 기울기

$x^3 -3x^2 + x$
- 접점에서의 기울기 = 접점에서의 미분계수 : $3x^2 - 6x + 1$
- if x == 0일때의 기울기 : 빨간 점에서의 기울기
- if x == 1일때의 기울기 : 녹색 점에서의 기울기

```python
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (6,4)

def f(x):
    return x**3 -3*x**2 + x

x = np.linspace(-1,3,400)
y = f(x)

plt.plot(x,y)
plt.plot(0,0, "ro")
plt.plot(x,x,"r:")
plt.plot(1,f(1), "go")
plt.plot(x,-2*(x-1)-1,"g--")


plt.grid()
plt.show()
```

```python
# 강사님 수업 code
def f(x):
    return x**3 -3 * x**2 + x

x = np.linspace(-1, 3, 400)
y = f(x)

plt.plot(x, y)
plt.plot(0, 0, 'ro')
plt.plot(x, x, 'r:')
plt.plot(1, -1, 'go')
plt.plot(x, (3*1**2-6*1+1)*(x-1)-1, 'g--')

plt.xlim(-3.5, 5.5)
plt.ylim(-4, 2)
plt.xticks(np.arange(-3, 6))
plt.yticks(np.arange(-4, 2))

plt.annotate('', xy=(1,0), xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=(1,1), xytext=(1,0), arrowprops=dict(facecolor='gray'))

plt.annotate('', xy=(2,-1), xytext=(1,-1), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=(2,-3), xytext=(2,-1), arrowprops=dict(facecolor='gray'))

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('함수의 기울기')
plt.grid()
plt.show()
```
