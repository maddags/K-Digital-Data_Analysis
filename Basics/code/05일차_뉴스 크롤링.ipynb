```python
import requests as req
from bs4 import BeautifulSoup
from datetime import datetime
import os
import pandas as pd

# 검색할 개수(원하는 개수 / 10개 할것)
count = 10
title_list = []
url_list = []
for i in range(count):
    url = "https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%AC%B4%ED%95%9C%EB%8F%84%EC%A0%84&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=42&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=" +str(i)
    res = req.get(url).text
    soup = BeautifulSoup(res,"html.parser")
# soup 
    title = soup.select(".news_area > a")
    for i in title :
        title_list.append(i.text)
        url_list.append(str(i.attrs["href"]))

data = {"기사제목" : title_list, "url" : url_list}
df = pd.DataFrame(data)
df
```
